```python
from types import resolve_bases
class Arr:
  def __init__(self, arr):
    self.arr = arr
  def alternateArr(self):
    res = []
    for i in range(0,len(self.arr),2):
      res.append(self.arr[i])
    return res
  def getArr(self):
    result = self.alternateArr()
    print(result)

arr1 = Arr([12,13,4,567,234,6])
arr1.getArr()
```

```python
class Leader:
  def __init__(self, arr):
    self.arr = arr
  def calculateLeader(self):
    res = []
    for i in range(0, len(self.arr)-1):
      if (self.arr[i] > self.arr[i+1]):
        res.append(self.arr[i])
    return res
  def getLeader(self):
    result = self.calculateLeader()
    print(result)
l1 = Leader([12,14,20,1,2,3,4,34,7])
l1.getLeader()
```
